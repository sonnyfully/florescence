# Convolution Reverb Wet Path

Stage 4 sanity plots for the fixed pre-convolution wet-path modulation and deterministic fallback IR. These mirror `Source/DSP/ConvReverbConfig.h`.

```python
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 48_000
rate_hz = 0.4
depth_percent = 0.002
center_delay_ms = 8.0
duration_seconds = 5.0

time = np.arange(int(sample_rate * duration_seconds)) / sample_rate
center_delay_samples = center_delay_ms * 0.001 * sample_rate
depth_samples = depth_percent * sample_rate
delay_samples = center_delay_samples + depth_samples * np.sin(2 * np.pi * rate_hz * time)

plt.figure(figsize=(10, 4))
plt.plot(time, delay_samples / sample_rate * 1000.0)
plt.xlabel("Time (s)")
plt.ylabel("Wet-path delay (ms)")
plt.title("Fixed pre-convolution modulation")
plt.tight_layout()
plt.show()
```

```python
fallback_length = 2048
decay = 0.997
early_gain = 0.35
samples = np.arange(fallback_length)
ir = 0.08 * (decay ** samples)
ir[0] = 1.0
ir[180] += early_gain

plt.figure(figsize=(10, 4))
plt.plot(samples / sample_rate * 1000.0, ir)
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude")
plt.title("Generated fallback IR shape")
plt.tight_layout()
plt.show()
```
